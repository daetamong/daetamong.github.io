```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
weight, bias = np.random.rand(2)

x = np.random.rand(50)

noise = np.random.uniform(-0.08, 0.08, size = x.shape)

y = weight * x + bias + noise
```

```python
plt.figure(figsize=(10, 5))
# plt.plot(x, y, color='r', label=f'y = {weight}*x + {bias}')
plt.scatter(x, y, label='data')
plt.show()
```

```python
# 각 파라미터의 초기값 설정
weight = np.random.uniform(low = -1.0, high = 1.0)
bias = np.random.uniform(low = -1.0, high = 1.0)
```

```python
y_hat = weight * x + bias
```

```python
MSE = (y_hat - y) ** 2
```

```python
learning_rate = 0.01
epoch = 1000
```

```python
errors = []
for epo in range(epoch):
    y_hat = weight * x + bias

    MSE = ((y_hat - y) ** 2).mean()
    if MSE < 0.0005:
        break

    weight = weight - learning_rate * ((y_hat - y) * x).mean()
    bias = bias - learning_rate * (y_hat - y).mean()

    errors.append(MSE)

    if epo % 5 == 0:
        print(f'weight : {weight} // bias : {bias} // MSE : {MSE}')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
